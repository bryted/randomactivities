Write a code to do the following:
1. Group the Methods of Payment into either Cash / Digital
2. Remove all Reversals("/") and their Corresponding Actual payments 
3. Remove all Jan 1 2021 cash deposits (NON-CASH)
4. Remove all withdrawals by username : admin from Jan - Dec 2021
5. Create Gender, Age and Age Group Columns and find their exact matches by Scheme Number from Member Data


```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('2020-2022PAY.xlsx')
df.head(100)
```

```python
# df["Year"] = df["[Sesseion date]"].dt.year
# df["Month"] = df["[Sesseion date]"].dt.month_name()
# df["Day"] = df["[Sesseion date]"].dt.day
```

```python
df['Product'].value_counts()
```

```python
df['Sector'] = df['Product'].replace({'PPT Personal Pension Scheme':'Informal','PPT Occupational Pension Scheme':'Formal','PPT TUC UNIWA Scheme':'Informal','PPT Provident Fund Scheme':'Formal'})
```

```python
df.Sector.value_counts()
```

```python
df
```

```python
# Creating a column based on 1st 3 letters of batch number
df['Group'] = df.iloc[:,3].str[:3]
```

```python
TR0 = df[df.Group == 'TR0']
PPT = df[df.Group == 'PPT']
IPA = df[df.Group == 'IPA']
PA0 = df[df.Group == 'PA0']
N209 = df[df.Group == '209']
N224= df[df.Group == '224']
```

```python
IPA['NewBatch'] = IPA.iloc[:,3].str[:14]
```

```python
PPT['NewBatch'] = PPT.iloc[:,3].str[:10]
```

```python
TR0['NewBatch'] = TR0.iloc[:,3].str[:12]
```

```python
frames = [PPT, IPA, TR0,PA0, N209,N224 ]
```

```python
result = pd.concat(frames)
```

```python
result
```

```python
result.NewBatch.fillna('Actual',inplace=True)
```

```python
result.Age.fillna(method='pad',inplace=True)
```

```python
result.isnull().sum()
```

```python
result['Reversal'] = result.iloc[:,3].str[-2:]
```

```python
result.loc[result['Reversal'] == '/1', 'Status'] = 'Reversal' 
result.loc[result['Reversal'] != '/1', 'Status'] = 'Actual' 
```

```python
result
```

```python
result.Status.value_counts()
```

```python
actual=result[result['Status']=='Actual']
```

```python
rever=result[result['Status']=='Reversal']
```

```python
C = actual.loc[~((actual.NewBatch.isin(rever["NewBatch"])))]
```
