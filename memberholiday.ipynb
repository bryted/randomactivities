```python
md = pd.read_excel('MembersDump (34).xlsx')
```

```python
md.info()
```

```python

md['Date of birth'] = pd.to_datetime(md['Date of birth'], errors ='coerce')
```

```python
md.info()
```

```python
md
```

```python
md['month'] = md['Date of birth'].dt.month
md['day'] = md['Date of birth'].dt.day
```

```python
feb = md[md['month']==2]
```

```python
feb
```

```python
feb1 = feb[['Gender', 'First name', '[Middle name]','[Last name]','[Scheme number]','Date of birth','Mobile','month','day']]
```

```python
feb1 = feb1.reset_index(drop = True)
```

```python
feb1
```

```python
#feb1.to_excel('feb1.xlsx')
```

```python
feb1['day'] = feb1[['day']].astype(int)
```

```python
feb1['day'] = feb1[['day']].astype(str)
```

```python
feb1
```

```python
days = feb1['day'].unique().tolist()
```

```python
days
```

```python
days.sort()
```

```python
days
```

```python
writer = pd.ExcelWriter( 'feb1.xlsx', engine='xlsxwriter')
for d in days:
  final = feb1.loc[feb1.day == d]
  final.to_excel(writer,sheet_name = d)
writer.save()
```
